```python
import pandas as pd
import re

df = filtered_df

def extract_tips(newick_str):
    # 匹配冒号前的所有非括号、非逗号内容
    candidates = re.findall(r"([^\(\),:]+):", newick_str)
    # 过滤掉纯数字（bootstrap/support值）
    tips = [c for c in candidates if not c.isdigit()]
    # 还可以去掉开头结尾的引号
    tips = [c.strip("'\"") for c in tips]
    return tips

all_tips = []
for tree_str in df['symbiont_subtree']:
    all_tips.extend(extract_tips(tree_str))

df_tips = pd.DataFrame(all_tips, columns=['microbe_tip'])

df_tips.shape
```

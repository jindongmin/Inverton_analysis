```python
import pandas as pd
from ete3 import Tree

# -----------------------------
# 1. 读取 metadata
# -----------------------------
meta_file = "invertible_counts_groups.tsv"
meta = pd.read_csv(meta_file, sep="\t")

# 使用 ID 列作为 tip
meta['tip'] = meta['ID'].str.strip().str.upper()

# group 列
meta['group'] = meta['group'].fillna(0).astype(str)  # 缺失 group 设置为 0

# 保留必要列
meta_small = meta[['tip','group']]

# -----------------------------
# 2. 读取树 tip
# -----------------------------
tree_file = "bac120_iqtree.nwk"
tree = Tree(tree_file, format=1)
tree_tips = [t.strip().upper() for t in tree.get_leaf_names()]

# -----------------------------
# 3. 找出树上缺失的 tip 并补齐
# -----------------------------
tips_missing = set(tree_tips) - set(meta_small['tip'])
if tips_missing:
    meta_extra = pd.DataFrame({
        'tip': list(tips_missing),
        'group': '0'
    })
    meta_clean = pd.concat([meta_small, meta_extra], ignore_index=True)
else:
    meta_clean = meta_small.copy()

# -----------------------------
# 4. 按树 tip 顺序排序
# -----------------------------
meta_clean = meta_clean.set_index('tip').reindex(tree_tips).reset_index()

# -----------------------------
# 5. 保存干净的 TSV
# -----------------------------
meta_clean.to_csv("invertible_counts_groups_clean.tsv", sep="\t", index=False)

# 查看前几行
meta_clean.head()
```

```python
q25 = df["inverton_number"].quantile(0.25)
q50 = df["inverton_number"].quantile(0.50)
q75 = df["inverton_number"].quantile(0.75)
labels = ["1", "2", "3", "4"]

df["group"] = pd.cut(
    df["inverton_number"],
    bins=[-float("inf"), q25, q50, q75, float("inf")],
    labels=labels
)
```
